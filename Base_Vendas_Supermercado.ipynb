```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('./base_de_dados/supermarket_sales.csv', encoding='utf-8', sep=',')
df
```

```python
df.info()
```

```python
df.columns = ['Fatura_ID','Filial','Cidade','Tipo de Cliente','Gênero','Linha de Produto','Preço Unitário','Quantidade Adquirida','Taxa 5% imposto','Total C/ Imposto','Data Venda','Hora Venda','Tipo Pagamento','Custo dos produtos','Porcentagem da margem bruta','Renda Bruta','Classificação']
```

```python
df
```

```python
df['Tipo de Cliente'].replace('Member', 'Membro', inplace=True)
df['Tipo de Cliente'].unique()
```

```python
df['Gênero'].replace('Female', 'Feminino', inplace=True)
df['Gênero'].replace('Male', 'Masculino', inplace=True)
df['Gênero'].unique()
```

```python
df['Linha de Produto'].replace('Health and beauty', 'Saúde e beleza', inplace=True)
df['Linha de Produto'].replace('Electronic accessories', 'Acessórios eletrônicos', inplace=True)
df['Linha de Produto'].replace('Home and lifestyle', 'Casa e estilo de vida', inplace=True)
df['Linha de Produto'].replace('Sports and travel', 'Esportes e viagens', inplace=True)
df['Linha de Produto'].replace('Food and beverages', 'Alimentação e bebidas', inplace=True)
df['Linha de Produto'].replace('Fashion accessories', 'Acessórios Fashions', inplace=True)
df['Linha de Produto'].unique()
```

```python
df['Tipo Pagamento'].replace('Ewallet', 'Carteira eletrônica', inplace=True)
df['Tipo Pagamento'].replace('Cash', 'Dinheiro', inplace=True)
df['Tipo Pagamento'].replace('Credit card', 'Cartão de Crédito', inplace=True)
df['Tipo Pagamento'].unique()
```

```python
conn = sqlite3.connect('Base_Vendas_Supermercado.db')
df.to_sql('table_vendas_supermercado', conn, if_exists='replace', index=False)
conn.close()
```
